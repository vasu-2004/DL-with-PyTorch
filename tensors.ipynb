```python
i2 = torch.tensor([[1,0],[0,1]])
m1 = torch.tensor([[1,2],[3,4]])
m2 = torch.tensor([[5,6],[7,8]])

prod1 = torch.mul(m1,m2) #element-wise
prod3 = m1*m2 #element-wise
prod2 = torch.mm(m1,m2) #matrix multiplication
prod4 = m1 @ m2

print(prod1)
print(prod3)
print(prod2)
print(prod4)
```

```python
import numpy as np

a = torch.ones(5)
print(a)
b = a.numpy()
print(b)
```

```python
if torch.cuda.is_available:
  device = torch.device("cuda")
  x = torch.ones(5,device=device)
  y = torch.ones(5)
  y = y.to(device)
  z = x+y
  print(z)
#the x+y oprn is performed on the GPU
```

```python
z_arr = z.numpy() #error- numpy can only handle CPU tensors
```

```python
z_arr = z.to("cpu").numpy()
z_arr
```

requires_grad=True is a property of tensors that tells PyTorch to track all operations on that tensor so that it can compute gradients during backpropagation. This is crucial in tasks involving optimization and training machine learning models.

```python
t = torch.ones(5,requires_grad=True)
print(t)
```
